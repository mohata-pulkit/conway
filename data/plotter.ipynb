```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
plt.rcParams["font.family"] = "BlexMono Nerd Font"
plt.rcParams["font.weight"] = "light"
```

```python
i = np.loadtxt("phys.txt", usecols=0)
num = np.loadtxt("phys.txt", usecols=1)
comx = np.loadtxt("phys.txt", usecols=2)
comy = np.loadtxt("phys.txt", usecols=3)
```

```python
dist = np.sqrt(comx**2 + comy**2)

plt.figure(figsize=(12,9))
plt.plot(i[0:50], dist[0:50], marker="o")
plt.savefig("dist.png", dpi=600, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12,9))
plt.plot(i[0:50], num[0:50], marker="o")
plt.savefig("num-alive.png", dpi=600, bbox_inches='tight')
plt.show()
```
